`split()`的作用就是将一串长的字符串通过分割符号进行分割

`while` 循环是一种在满足特定条件时重复执行代码块的控制结构，你可以通过设置`while`循环的条件部分为真(布尔值`True`)， 这样，循环将一直执行，直到你使用`break`语句来终止循环。

`try`:尝试执行`try`中的代码，如果程序异常或者发生错误，则由`except`捕获异常并执行异常处理代码

```python
while True:
    try:
        data = input().split()
        # 这里将列表data里的元素转换为int类型，然后相加
        res = int(data[0]) + int(data[1])
        print(res)
    except:
        break
```
