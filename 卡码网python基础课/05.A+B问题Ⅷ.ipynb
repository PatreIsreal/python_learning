```python
while True:
    try:
        N = int(input())
        for i in range(N):
            inputs = input().split()
            M = int(inputs[0])
            total = 0
            for j in range(M):
                total += int(inputs[j+1])
            print(total)
        
        if i < N-1:
            print()
    except:
        break
```

- `sum()``map()`

```python
while True:
    try:
        N = int(input())
        for i in range(N):
            line = list(map(int, input().spilt()))
            M = line[0]
            total = sum(line[1:])
            print(total)

        if i < N-1:
            print()

    except:
        break
```
