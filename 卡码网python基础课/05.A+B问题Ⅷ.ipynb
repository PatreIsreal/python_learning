- 第一种方法：遍历累加

```python
while True:
    try:
        N = int(input())
        for i in range(N):
            inputs = input().split()
            M = int(inputs[0])
            total = 0
            for j in range(M):
                total += int(inputs[j+1])
            print(total)
        
        if i < N-1:
            print()
    except:
        break
```

- 第二种解法，使用`sum()`和`map()`函数进行求解

```python
while True:
    try:
        N = int(input())
        for i in range(N):
            line = list(map(int, input().spilt()))
            M = line[0]
            total = sum(line[1:])
            print(total)

        if i < N-1:
            print()

    except:
        break
```
