`upper()`

````python
n = int(input())  # 

for _ in range(n):
    s = input()  # 
    result = ''

    # 
    if s[0].islower():
        result += s[0].upper()
    else:
        result += s[0]

    # 
    for i in range(1, len(s) - 1):
        # 
        if s[i] == ' ' and s[i + 1] != ' ':
            # 
            if s[i + 1].islower():
                result += s[i + 1].upper()
            else:
                result += s[i + 1]

    print(result)
````

### `split()`

````python
n = int(input())
for _ in range(n):
    words = input().split()
    result = ''
    # 
    for word in words:
       # result
        result += word[0].upper()
    print(result)
````
