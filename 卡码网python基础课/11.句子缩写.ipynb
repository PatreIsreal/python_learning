Python提供了对应的方法`islower()`用于检查字符串中的字符是否都是小写字母，因此可以使用`s[0]`获取第一位字符，然后检查是否是小写字母。

转换成大写字符可以使用`upper()`方法，用于将字符串中的所有字符转换为大写字母。

````python
n = int(input())  # 输入测试用例的数量

for _ in range(n):
    s = input()  # 输入一行字符串
    result = ''

    # 如果首字母是小写，转换成大写
    if s[0].islower():
        result += s[0].upper()
    else:
        result += s[0]

    # 遍历整个字符串
    for i in range(1, len(s) - 1):
        # 如果当前字符是空格，下一个字符不是空格，说明下一个字符是新单词的首字母
        if s[i] == ' ' and s[i + 1] != ' ':
            # 判定新单词的首字母是否是小写，小写则转换成大写
            if s[i + 1].islower():
                result += s[i + 1].upper()
            else:
                result += s[i + 1]

    print(result)
````

### 简易写法：利用`split()`

````python
n = int(input())
for _ in range(n):
    words = input().split()
    result = ''
    # 遍历单词列表
    for word in words:
       # 取得每个单词的首字母，将之转为大写，然后拼接到result中
        result += word[0].upper()
    print(result)
````


### 代码解析
#### 1. 读取总行数
```python
num = int(input().strip())
```
- **`input().strip()`**：读取用户输入的一行内容，并**去除首尾的空白字符**（如换行符、多余空格）。
- **`int()`**：将输入转换为整数，表示后续需要处理的行数。

#### 2. 循环处理每一行
```python
for i in range(num):
    # 处理单行并输出结果
    print(''.join([j[0].upper() for j in input().strip().split()]))
```
- **`input().strip().split()`**：
  - **`strip()`**：去除当前行的首尾空白。
  - **`split()`**：默认以**任意连续空白符（空格、制表符等）**分割字符串，得到一个单词列表。例如 `"  hello   world  "` 会转换为 `["hello", "world"]`。
- **列表推导式 `[j[0].upper() ...]`**：
  - **`j[0]`**：取每个单词的第一个字符。
  - **`upper()`**：将字符转换为大写，确保缩写字母统一为大写。
- **`''.join(...)`**：将列表中的所有首字母拼接成一个字符串（如 `['H', 'W']` → `"HW"`）。

### 为什么更简洁？
与原代码相比，这段代码通过以下优化大幅简化逻辑：

#### 1. **利用 `split()` 自动处理分词**
- **原代码**：手动遍历字符串，检查空格并提取下一个字符作为新单词首字母。
- **新代码**：直接通过 `split()` 分割出单词列表，无需处理空格的复杂逻辑，代码更健壮（如多个空格、首尾空格等场景）。

#### 2. **直接操作单词首字母**
- **原代码**：逐个字符判断是否为新单词首字母，需要处理边界条件（如首字符、末尾字符）。
- **新代码**：直接遍历分割后的单词列表，取每个单词的首字符，逻辑更直观。

#### 3. **使用列表推导式替代显式循环**
- **原代码**：通过 `for` 循环逐字符处理，手动拼接结果。
- **新代码**：用一行列表推导式完成所有单词首字母的提取和转换，再通过 `join` 快速拼接结果。


### 更简洁的写法：

````python
num = int(input().strip()) #总共几行
for i in range(num):
    print(''.join([j[0].upper() for j in input().strip().split()]))
````
