- 考察循环嵌套的知识点

>四条边界上输出*，中间中空的地方输出" "

```python
# 这种是按行输出，只有第一列和最后一列全是*
# 没有循环嵌套，下面的代码按照循环嵌套来做
while True:
    try:
        n = int(input())
        for i in range(n):
            if i == 0 or i == n-1:
                print("*"*n)
            else:
                print("*" + " "*(n-2) + "*")
    except:
        break
```

```python
# 这里按照循环嵌套来做，行和列同时索引，一个元素一个元素的输出
# 四个边输出*，其他位置输出空格
while True:
    try:
        n = int(input())
        for i in range(n):
            for j in range(n):
                if i == 0 or i == n-1 or j == 0 or j == n-1:
                    print("*", end="")
                else:
                    print(" ", end="")
            print()
    except:
        break
```
