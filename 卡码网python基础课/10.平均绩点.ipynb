- 字符串遍历，使用`for`循环

- 使用flag方法控制程序的输出

```python
while True:
    try:
        s = list(input().split())
        n = len(s)
        flag = True #用flag控制程序是否输出结果
        total = 0
        for i in s:
            if i == 'A':
                total += 4
            elif i == 'B':
                total += 3
            elif i == 'C':
                total += 2
            elif i == 'D':
                total += 1
            elif i == 'F':
                total += 0

            else:
                print('Unknown')
                flag = False # 这里将flag设置为False，表示不输出结果，不执行下边的那个if
                break

        if flag:
            average = total / n
            print(f'{average:.2f}')
    except:
        break

```
