```python
while True:
    try:
        s = list(input().split())
        n = len(s)
        flag = True #flag
        total = 0
        for i in s:
            if i == 'A':
                total += 4
            elif i == 'B':
                total += 3
            elif i == 'C':
                total += 2
            elif i == 'D':
                total += 1
            elif i == 'F':
                total += 0

            else:
                print('Unknown')
                flag = False # flagFalseif
                break

        if flag:
            average = total / n
            print(f'{average:.2f}')
    except:
        break

```
