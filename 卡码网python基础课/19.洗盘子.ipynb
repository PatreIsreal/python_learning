- 列表模仿桩的使用
- >对于待清洗盘子来说是出栈，对于已清洗区域来说，是入栈（进栈）



- 如果一个列表为空，`not stack`将会返回`True`，表示列表为空；如果列表非空，`not stack`将会返回`False`，表示列表不为空。

```python
# 完整代码
# 读取输入依旧使用拆分的方法以列表的形式表示
n = int(input())
plates = list(map(int, input().split()))

# m次操作
m = int(input())
for _ in range(m):
    # 获取对盘子的操作，如果是第二种操作，有两个数字，所以拆分
    operations = input().split()
    # 第一个数字是进行了什么操作
    opt = int(operations[0])
    # 操作1：拿走并清洗盘子
    if opt == 1:
        # 要先确定是不是空栈
        if plates:
            cleaned = plates.pop()

    elif opt == 2:
        # 盘子的编号是输入的开头是2的那一行的第二个数字
        plate_n = int(operations[1])
        # 向栈中添加盘子
        plates.append(plate_n)
# 判断堆放盘子区是否为空
if not plates:
    print("All the dishes have been washed.")
else:
    print(plates[-1])
```
