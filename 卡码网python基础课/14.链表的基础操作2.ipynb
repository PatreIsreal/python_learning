**定义链表和链表节点，以及链表插入和打印链表的函数，这些内容都是“模板”写法**

```python
class Node:
    def __init__(self, data):
        self.data = data  # 节点数据
        self.next = None  # 指向下一个节点的指针，初始为None

class LinkList:
    def __init__(self):
        self.head_node = None  # 链表的头节点，初始为None
        self.length = 0  # 链表长度，初始为0

    def insert(self, data):
        self.length = self.length + 1  # 增加链表长度
        new_node = Node(data)  # 创建新节点

        if self.head_node is None:
            self.head_node = new_node  # 如果链表为空，新节点成为头节点
            return self.head_node
        else:
            current_node = self.head_node  # 从头节点开始遍历

            while current_node.next is not None:  # 这里有逻辑错误，应该是while current_node.next is not None
                current_node = current_node.next  # 移动到下一个节点
            current_node.next = new_node  # 将新节点连接到链表末尾

            return new_node  # 返回新节点
```

### 也可以将寻找第n个元素写成方法

>类似于嵌入方法一样

**完整代码如下**：


```python
# 链表节点定义
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# 链表定义
class LinkList:
    # 初始化
    def __init__(self):
        self.head_node = None
        self.length = 0
    # 定义嵌入方法
    def insert(self, data):
        self.length = self.length + 1
        new_node = Node(data)
        # 如果链表头节点是空，那么将这个节点设置为头节点
        if self.head_node is None:
            self.head_node = new_node
            return self.head_node
        # 如果不是空的
        else:
            # 从头节点开始遍历
            current_node = self.head_node
            # 所遍历到的节点的下一个节点不是空的
            # 就一直往下遍历
            while current_node.next is not None:
                current_node = current_node.next # 往下遍历
            current_node.next = new_node # 直到遍历到最后一个，也就是空节点
            # 返回这个输入的节点
            return new_node
    # 获取第n个节点的方法
    def get(self, n):
        # 如果输入位置不合法
        if n < 1 or n > self.length:
            return None
        i = 1
        # 从头开始遍历
        current_node = self.head_node
        # 目前遍历到的节点不是空的时候
        while current_node is not None:
            if i == n:
                # 遍历到了就返回current_node
                return current_node

            i = i + 1
            current_node = current_node.next
        return None
# 输入链表长度n和取出元素的数量k
n, k = map(int, input().split())
# 从LinkList类创建链表
link_list = LinkList()
# 输入链表的元素，将其转换为列表的整数形式
elements = list(map(int, input().split()))
# 调用insert方法，往链表中输入元素
for data in elements:
    link_list.insert(data)

# 输入需要取出的元素的位置的数字
m_values = list(map(int, input().split()))

for m in m_values:
    # 调用get方法，从链表中取出第m个元素
    node = link_list.get(m)
    # 将取出的node打印出来
    # 如果节点不为None，打印出来
    if node is not None:
        print(node.data)
    # 如果节点是None，打印这段语句
    else:
        print("Output position out of bounds.")
```
