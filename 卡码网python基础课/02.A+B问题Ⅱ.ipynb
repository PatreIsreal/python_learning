- `range()`函数

```python
# 只指定 stop, 生成 0 - 4的数字
for i in range(5):
    print(i)

# 指定start, stop, 生成 2 - 5的数字
for i in range(2, 6):
    print(i)

# 指定起始值、终止值和步长， 生成 1 3 5 7 9
for i in range(1, 10, 2):
    print(i)
```

- 获取整数N，让下面的程序执行N次

```python
while True:
    try:
        N = int(input("请输入一个整数: "))
        for _ in range(N):
            data = input().split()
            res = int(data[0]) + int(data[1])
            print(res)
    except: 
        break
    
```

- 也可以用`while`循环写

```python
while True:
    try:
        N = int(input())
        # 初始化一个值 0
        i = 0
        # 判断 i 是否 小于 N 
        while i < N:
            a, b = input().split()
            # 计算结果
            print(int(a) + int(b))
            # 将 i 的值累加 1
            i = i + 1
    except:
        break
```
