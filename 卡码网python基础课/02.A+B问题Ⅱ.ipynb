- `range()`

```python
#  stop,  0 - 4
for i in range(5):
    print(i)

# start, stop,  2 - 5
for i in range(2, 6):
    print(i)

#   1 3 5 7 9
for i in range(1, 10, 2):
    print(i)
```

- NN

```python
while True:
    try:
        N = int(input(": "))
        for _ in range(N):
            data = input().split()
            res = int(data[0]) + int(data[1])
            print(res)
    except: 
        break
    
```

- `while`

```python
while True:
    try:
        N = int(input())
        #  0
        i = 0
        #  i   N 
        while i < N:
            a, b = input().split()
            # 
            print(int(a) + int(b))
            #  i  1
            i = i + 1
    except:
        break
```
