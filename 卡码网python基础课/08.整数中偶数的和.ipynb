- : `//`

```python
while True:
    try:
        n = int(input())
        result = 0
        while n != 0: # n0
            a = n % 10 # 
            n = n // 10 # 

            if a % 2 == 0: # 
                result += a
    
        print(result)
        print(" ")
    except:
        break
```
