- 取余操作：`%`

- 取整操作: `//`

```python
while True:
    try:
        n = int(input())
        result = 0
        while n != 0: # n不等于0的时候，一直往下除
            a = n % 10 # 取余数
            n = n // 10 # 取整数

            if a % 2 == 0: # 如果余数是偶数，就加上
                result += a
    
        print(result)
        print(" ")
    except:
        break
```
