## 列表

- 创建列表：
    - 使用方括号 []或使用 list() 创建一个列表。
- 访问列表元素：
    - 通过索引来进行访问，注意是从0开始的，[0]代表列表中的第一个元素
- 修改列表元素：
    - 通过索引来进行修改
- 列表长度：
    - 使用`len()`来获取列表的长度

### 操作列表的方法：

1. `append(value)`：将新的元素添加到列表的末尾。

2. `insert(index, value)`: 在指定位置插入元素。

3. `remove(value)`: 移除值为value的元素，但是如果有多个时，只删除第一个。

4. `pop(index)`: 删除并返回指定索引index位置的元素。

5. `index(value)`: 返回指定值value的索引位置, 如果有多个时，返回第一个。

6. `sort()`: 按照升序排序列表。

7. `reverse()`: 反转列表的顺序。

### 切片

```python
my_list = [1, 2, 3, 4, 5, 6, 7]

# 从索引1到索引5，不包括索引5的元素
sublist = my_list[1:5]
# sublist 现在是 [2,3,4,5]

# 从索引1开始到末尾
sublist = my_list[1:]
# sublist 现在是 [2, 3, 4, 5, 6, 7]

# 从头开始到索引3，不包括索引3的元素
sublist = my_list[:3]
# sublist 现在是 [1, 2, 3]

# 使用步长，从索引0开始，每隔2个元素取一个
sublist = my_list[0::2]
# sublist 现在是 [1，3，5，7]
```

>此外，还可以使用负数索引，-1 表示最后一个元素，-2表示倒数第二个元素，依次类推。

## 解题

```python
n = int(input()) # 读取输入的整数，这是下一步输入的数的个数
line = list(map(int, input().split())) # 读取输入的整数，以空格分隔，转换为整数列表
print(line) # 输出整数列表
# 起始值n - 1，终止值-1，步长-1,这里不包括终于值-1，所以到0，即第一个元素，是开区间
for i in range(n - 1, -1, -1): # 从最后一个元素到第一个元素，步长为-1
    print(line[i], end=' ')
print( )
for i in range(0, n, 2):
    print(line[i], end=' ')
```
