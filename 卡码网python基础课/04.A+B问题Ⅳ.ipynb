```python
while True:
    try:
        inputs = input().split()
        if inputs[0] == '0':
            break
        total = 0
        for i in range(int(inputs[0])):
            total += int(inputs[i+1])
        print(total)

    except:
        break
```

```python
while True:
    data = input().split()
    n = int(data[0])
    if n == 0:
        break
    numbers = list(map(int, data[1:]))
    total = sum(numbers)
    print(total)
    
```
