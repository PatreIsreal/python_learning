```python
while True:
    try:
        inputs = input().split()
        if inputs[0] == '0':
            break
        total = 0
        for i in range(int(inputs[0])):
            total += int(inputs[i+1])
        print(total)

    except:
        break
```

- 加入`sum()`函数和`map()`函数
- `map()`可以将一个函数应用到序列的每个元素，并返回一个包含结果的新序列。
- `input_line[1:]`是一个切片操作，表示从列表中获取从指定索引（包括该索引）到序列末尾的所有元素, 然后对这些元素都执行`int`函数转换，将之转换成整数

```python
while True:
    data = input().split()
    n = int(data[0])
    if n == 0:
        break
    numbers = list(map(int, data[1:]))
    total = sum(numbers)
    print(total)
    
```
