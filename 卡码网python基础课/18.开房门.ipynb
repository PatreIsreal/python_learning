### 用映射(map)实现哈希表

我们常常把**map**称之为映射，就是将一个元素（通常称之为`key`）与一个相对应的值（通常称之为`value`）关联起来，比如说一个学生的姓名（`key`）有与之对应的成绩（`value`），它们是一一对应的，就好像一把钥匙打开一扇门，在**map**中`key`是唯一的，也只有一个唯一的确定的值。

比如下面的图示：左边是 `key`，每一个姓名都对应着一个成绩 `value`。


1. 这种就是字典的创建
```python
my_dict = {
  "mike", 99,
  "tom": 98,
  "jerry": 100
}
```
- 键的数据类型是不可变的，但是值是可以任意变得



>如果直接创建一个{}, 表示这是一个空字典，而并非一个空集合。

2. 用dict()创建字典
```python
# 创建一个空字典
my_dict = dict()
# 添加键值对
my_dict["mike"] = 99
my_dict["tom"] = 98
my_dict["jerry"] = 100

# 更新值
my_dict["mike"] = 100
# 访问值
print(my_dict["mike"]) # 100
```

- 删除字典中的键值用del()或者clear()方法

```python
del my_dict["mike"] # 删除对应的键值对
my_dict.clear() # 清空字典
del my_dict # 删除字典
```

- 检查字典中的键是否存在也用in来检测

```python
if "mike" in my_dict:
      print(my_dict["mike"])
else:
      print("mike doesn't exist in the dictionary")

```

- 遍历字典的数据

```python
for key, value in my_dict.items():
      print(f"Key: {key}, Value: {value}")

```

````python
# 完整代码如下：
# 依旧是几组数据
k = int(input())
for _ in range(k):
    # 先创建一个空字典
    ump = {}
    # 读取键值对的数量
    n = int(input())

    for _ in range(n):
        # 接收输入的两个整数分别是钥匙和门，进行一个拆分
        key, door = map(int, input().split())
        # 将钥匙和门一一填入字典
        ump[key] = door

    # 读取要查询的门的编号
    x = int(input())
    flag = True
    # 遍历字典中的键和值
    for key, door in ump.items():
        if door == x: # 如果找到对应的门
            print(key) # 输出对应的键
            # 找到并且输出对应的键之后，将flag变为False
            flag = False
            break
    # 如果没有遍历完了都没有找到，这个时候没有执行上边的break，没有跳出循环
    # 这个时候flag是True
    if flag:
        print("Can't open the door.")
````
