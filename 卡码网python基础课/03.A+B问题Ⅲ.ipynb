- 前面正常输入输出，当输入的两个数字都为0时，不进行加法，系统跳出循环，结束运行

```python
while True:
    try:
        data = input().split()
        a, b = int(data[0]), int(data[1])
        if a == 0 and b == 0:
            break
        print(a + b)
    except:
        break
```

- `break`和`continue`的比较

`break`是停止循环，`continue`是不执行当前循环的剩余部分，进行下一次循环

```python
# 这里用continue，会跳出这一次循环，不会进行计算0+0，直接进入下一次循环,但不会终止循环，还是会继续输入
while True: 
    try:
        s = input().split()
        a, b = int(s[0]), int(s[1])
        if a == 0 and b == 0:
            continue
        print(a + b)
    except:
        break
```
