```python
while True:
    try:
        data = input().split()
        a, b = int(data[0]), int(data[1])
        if a == 0 and b == 0:
            break
        print(a + b)
    except:
        break
```

- `break``continue`

`break``continue`

```python
# continue0+0,
while True: 
    try:
        s = input().split()
        a, b = int(s[0]), int(s[1])
        if a == 0 and b == 0:
            continue
        print(a + b)
    except:
        break
```
