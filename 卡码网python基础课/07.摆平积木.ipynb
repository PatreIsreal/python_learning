```python
while True:
    try:
        n = int(input())
        # 输入为0时结束
        if n == 0:
            break

        numbers = list(map(int, input().split()))
        total = sum(numbers)
        # 保证能整除，// 为整除
        average = total // n
        # 这里必须要初始化为0，否则会出现错误
        times = 0
        for i in numbers:
            # 如果大于平均值，就加上差值，算到总次数里
            if i > average:
                times += i - average
        print(times)

    except:
        break
```
