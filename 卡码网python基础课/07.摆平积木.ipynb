```python
while True:
    try:
        n = int(input())
        # 0
        if n == 0:
            break

        numbers = list(map(int, input().split()))
        total = sum(numbers)
        # // 
        average = total // n
        # 0
        times = 0
        for i in numbers:
            # 
            if i > average:
                times += i - average
        print(times)

    except:
        break
```
