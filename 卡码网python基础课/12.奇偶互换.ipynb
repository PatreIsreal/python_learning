````python
# 
def swap(a, b):
    return b, a

n = int(input())
for _ in range(n):
    s = input()

# 
    chas = list(s)
# 
    for i in range(0, len(chas)-1, 2):
        chas[i], chas[i+1] = swap(chas[i], chas[i+1])

    print("".join(chas))
````
