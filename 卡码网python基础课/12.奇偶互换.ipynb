````python
# 定义交换位置的函数
def swap(a, b):
    return b, a

n = int(input())
for _ in range(n):
    s = input()

# 将输入的字符串转换为字典型列表
    chas = list(s)
# 从第一位开始遍历，每次跳两位，交换该位置和下一位的值
    for i in range(0, len(chas)-1, 2):
        chas[i], chas[i+1] = swap(chas[i], chas[i+1])

    print("".join(chas))
````
